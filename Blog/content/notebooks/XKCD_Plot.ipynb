In order to create the XKCD_plot, several things needs to be done on Ubuntu 14.04.
0. You need to have matplotlib and make sure it works.
1. Install the necessary font: Humor Sans from the following link
https://github.com/shreyankg/xkcd-desktop/blob/master/Humor-Sans.ttf
And just intsall it as usuall font.
2. Remove the .cache file, ~/.cache/matplotlib/fontList.cache, of matplotlib.

```python
plt.xkcd()  # Yes...
fig=plt.figure()
plt.plot(sin(linspace(0, 10)))
plt.title('This is a wicked sin!!!')

fig.savefig('sin.pdf',bbox_inches='tight')# The pdf file output is correct!
```
